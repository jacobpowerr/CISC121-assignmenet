```python
class Solution(object):
    def isPalindrome(self, s):
        s == s.strip().lower()
        return s == s[::-1]
```

# 3. Final Correct Code

```python
# Used Chatgpt to hint to me why my code was not passing test
# Example prompt: give me some hints as to why my code is failing
class Solution(object):
    def isPalindrome(self, s):
        cleaned = "".join(ch.lower() for ch in s if ch.isalnum())
        return cleaned == cleaned[::-1]
```
